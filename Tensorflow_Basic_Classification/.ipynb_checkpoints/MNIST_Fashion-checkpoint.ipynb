```python
import tensorflow as tf
from tensorflow import keras

import numpy as np
import matplotlib.pyplot as plt

print (tf.__version__)
```

```python
fashion_mnist = keras.datasets.fashion_mnist

(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
```

# --------RUN CELLS BELOW THIS LINE--------

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover',
               'Dress', 'Coat', 'Sandal', 'Shirt',
               'Sneaker', 'Bag', 'Ankle boot']
```

## Visualize data

#### Training Images

```python
# View training images
print(train_images.shape)

# View first image
print(train_images[1,:10,:10])
```

#### Training Labels

```python
# View shape of the training labels
print(train_labels.shape)

# Check the first ten labels
print(train_labels[:10])
```

### Preprocess Data

```python
plt.figure()
plt.imshow(train_images[0], cmap='binary')
plt.colorbar()
```

```python
train_images = train_images/255.0
test_images = test_images/255.0
```

```python
# Plot first testing image
plt.figure()
plt.imshow(train_images[0], cmap='binary')
plt.colorbar()
# Plot first training image
plt.figure()
plt.imshow(test_images[0], cmap='binary')
plt.colorbar()
```

```python
plt.figure(figsize=(8,8))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(train_images[i], cmap=plt.cm.binary)
    plt.xlabel(class_names[train_labels[i]])

```

### Building the Model !!

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(128, activation=tf.nn.relu),
    keras.layers.Dense(10,activation=tf.nn.softmax)
])
```

```python
model.compile(optimizer=tf.train.AdamOptimizer(),
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])
```

```python
model.fit(train_images, train_labels, epochs=5)
```

```python
predictions = model.predict(test_images)
```

```python
first_pred = np.argmax(predictions[0])
print(predictions[0], '\n')
print('{}: {}'.format(first_pred, class_names[first_pred]), '\n')

plt.figure()
plt.imshow(test_images[0])
```

```python
plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(test_images[i], cmap=plt.cm.binary)
    # Find true and predicted labels
    true_label = test_labels[i]
    predicted_label = np.argmax(predictions[i])
    # Change color if correct or incorrect
    if predicted_label == true_label:
        color = 'green'
    else:
        color = 'red'
    # Plot x label
    plt.xlabel('{} ({})'.format(class_names[predicted_label],class_names[true_label]),
              color=color)
    
```

```python
x = 17

test_img = test_images[x]
print(test_img.shape)
test_img = np.expand_dims(test_img, 0)
print(test_img.shape)
```

```python
y = np.argmax(model.predict(test_img)[0])
yhat = test_labels[x]

print('{}, {}'.format(y,    class_names[y]   ))
print('{}, {}'.format(yhat, class_names[yhat]))
print('\nEqual? \n{}!'.format(y==yhat))
```
