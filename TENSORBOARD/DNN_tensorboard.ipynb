````python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import pandas as pd



print (tf.__version__)
````

````python
tf.enable_eager_execution()
````

````python
boston_housing = keras.datasets.boston_housing

(train_data, train_labels), (test_data, test_labels) = boston_housing.load_data()
````

````python
column_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD',
                'TAX', 'PTRATIO', 'B', 'LSTAT']

df = pd.DataFrame(data=train_data, columns=column_names)
df.head()
````

````python
df.describe()
````

````python
df.corr()
````


**Attributes**
- CRIM:    per capita crime rate by town.
- ZN:      proportion of residential land zoned for lots over 25,000 sq.ft.
- INDUS:   proportion of non-retail business acres per town.
- CHAS:    Charles River dummy variable (= 1 if tract bounds river; 0 otherwise).
- NOX:     nitric oxides concentration (parts per 10 million).
- RM:      average number of rooms per dwelling.
- AGE:     proportion of owner-occupied units built prior to 1940.
- DIS:     weighted distances to ﬁve Boston employment centers.
- RAD:     index of accessibility to radial highways.
- TAX:     full-value property-tax rate per 10,000 dollars.
- PTRATIO: pupil-teacher ratio by town.
- B:       1000(Bk−0.63)^2 where Bk is the proportion of blacks by town.
- LSTAT:   Percent lower status of the population.
- MEDV:    Median value of owner-occupied homes in thousand dollars.

````python
mean = np.array(df.mean())
std = np.array(df.std())
````

````python
x_train = (train_data - mean) / std
x_test = (test_data - mean) / std
````

````python
df_norm = pd.DataFrame(x_train, columns=column_names)
df_norm['Y'] = train_labels
df_norm.head()
````

````python
df_norm_test = pd.DataFrame(x_test, columns=column_names)
df_norm_test['Y'] = test_labels
````

````python
my_fc = [tf.feature_column.numeric_column(x) for x in df_norm.keys()]
print (my_fc)
````

````python
estimator = tf.estimator.DNNRegressor(
    hidden_units=[32,16],
    feature_columns=my_fc,
    model_dir='./DNN',
    weight_column=None,
    optimizer='Adagrad',
    activation_fn=tf.nn.relu,
    dropout=0.25,
    input_layer_partitioner=None,
    config=None
)
````

````python
# make the input_fn for the training
input_fn_train = tf.estimator.inputs.numpy_input_fn(
      x=dict(df_norm),
      y=np.array(train_labels),
      num_epochs=None,
      shuffle=True)
````

````python
# make the input_fn for the testing
input_fn_test = tf.estimator.inputs.numpy_input_fn(
      x=dict(df_norm_test),
      y=np.array(test_labels),
      num_epochs=1,
      shuffle=True)
````

````python
train_spec = tf.estimator.TrainSpec(input_fn=input_fn_train,
                                    max_steps=5000)
eval_spec = tf.estimator.EvalSpec(input_fn=input_fn_test)
````

````python
tf.estimator.train_and_evaluate(estimator=estimator,
                                train_spec=train_spec,
                                eval_spec=eval_spec
                                )
````
